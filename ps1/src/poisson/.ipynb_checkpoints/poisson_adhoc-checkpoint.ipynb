```python
## single step batch gradient descent

def batch_gradient_step(theta, x, y, step_size):
    
    
    # looping through each theta dimension
    for j in range(len(theta)):
        
        #looping through each point to update the above theta value
        for i in range(len(y)):
            # hypothesis = math.exp(x.dot(np.atleast_2d(theta).T)[i][0])
            hypothesis = math.exp((x @ theta)[i])
            
            theta[j] = theta[j] + step_size*(y[i] - hypothesis)*x[i][j]
            
            
    return theta       
            
        
```

```python
## single step another way -> very fast

def batch_gradient_step(theta, x, y, step_size):
    
    difference_vec = y - np.exp(x @ theta)
    
    # multiply it with x transpose. intuitive!
    
    theta_step = step_size * (x.T @ difference_vec)
            
    return theta + theta_step  
```

```python
## loss would the MSE between y_true and hypothesis function

def loss(theta, x, y):
    
    # hypothesis = np.exp(x.dot(np.atleast_2d(theta).T).T[0])
    hypothesis = np.exp(x @ theta)
    loss_vec = y - hypothesis
    loss = np.linalg.norm(loss_vec)
    return loss
    
```

```python
step_size=2e-7
max_iter=10000000 
eps=1e-5
verbose=True
theta_0 = None
```

```python
if theta_0 is None:
    theta = np.zeros(x_train.shape[1])
```

```python
theta
```

```python
iter_count = 1

initial_loss = loss(theta, x_train, y_train)

print('Initial loss value is: {0}'.format(initial_loss))

while(iter_count<=max_iter):
    
    # storing the initial theta value, need to create a copy not a reference
    theta_prev = theta.copy()
    
    # updating theta
    theta = batch_gradient_step(theta, x_train, y_train, step_size=step_size)
    
    # Loss value calc
    
    loss_updated = loss(theta, x_train, y_train)
    
    print('Loss after {0} iteration is: {1}'.format(iter_count, loss_updated))
    
    # step size norm change
    
    delta_theta_norm = np.linalg.norm(theta - theta_prev)
    
    print('Norm change in parameters is: {0}'.format(delta_theta_norm))
    
    # Stopping criterion
    if delta_theta_norm < 1e-5:
        break
    
    # forwarding the loop
    iter_count = iter_count+1
    
    
```
